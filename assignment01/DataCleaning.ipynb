```python
import pandas as pd
import numpy as np
```

### Load Data

Download Data here: https://gtexportal.org/home/downloads/adult-gtex/bulk_tissue_expression



```python
Gene = pd.read_csv("GTEx_Analysis_20170605_v8_RNASeQCv1.1.9_gene_tpm.gct",sep ="\t",skiprows=2,engine="c")
```

```python
Gene.head()
```

```python
Meta = pd.read_csv("Meta.txt",sep ="\t")
Meta.head()
```

### Get Top 10 tissues

```python
#get top 10 tissues
print(len(Meta))
t=Meta["SMTS"].value_counts().head(10).index
print(t)
Meta = Meta[Meta["SMTS"].isin(t)]
print(len(Meta))
```

### Get genes and calculate variance 



```python
geneVals = Gene.iloc[:,2:]
geneVals.head()
```

```python
geneVals_Array = geneVals.to_numpy()
gene_variance = np.nanvar(geneVals_Array,axis=1)
```

```python
gene_variance
```

```python
Gene["gene_variance"]=gene_variance
Top5kGenes = Gene.sort_values(by="gene_variance",ascending=False,kind="mergesort").head(5000)
Top5kGenes.head()
```

#### Create intermediate Top5kGenes file

```python
#Top5kGenes.to_csv("Top5kGenes.csv",index=False)
Top5kGenes = pd.read_csv("Top5kGenes.csv")
```

```python
Top5kGenes.head()
```

Transpose dataframe, make first gene row the columns and drop the description row

Map the tissue to the sample name

```python
gene_features=Top5kGenes.transpose().reset_index()
```

```python
gene_feats = gene_features.drop(gene_features.index[0]) 
gene_feats.columns = gene_features.iloc[0]#make column names the gene name
gene_feats = gene_feats.drop(gene_feats.index[0])#remove the gene description row
gene_feats.head()
```

```python
gene_feats["tissue"].unique()
```

```python
samp_tissue = dict(zip(Meta["SAMPID"],Meta["SMTS"]))
gene_feats["tissue"] = gene_feats["Name"].map(samp_tissue)
gene_feats.head()
```
